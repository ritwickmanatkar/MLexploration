```python
drop_columns = ['Photo', 'Flag','Club Logo',  
                'Real Face', 'Jersey Number', 'Loaned From',
                'Height', 'Weight','LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
                'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM',
                'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB', 'Release Clause']

df_1 = df.copy()
df_1.drop(drop_columns, axis=1, inplace=True)
df_1.head()
```

```python
df_1.shape
```

## Fixing the columns we need : 

```python
def convert(string):    
    if string[-1] == 'M':
        return float(string[1:-1])*1
    elif string[-1] == 'K':
        return ((float(string[1:-1])*1000) / 1000000)
    else:
        return float(string[1:])
    
df_2 = df_1.copy()
df_2['Value_in_Euros_in_M'] = df_2['Value'].apply(lambda x: convert(x))
df_2['Wage_in_Euros_in_M'] = df_2['Wage'].apply(lambda x: convert(x))
df_2.drop(['Value', 'Wage'], axis = 1, inplace = True)
df_2.head()
```

## Descriptive Analysis of the Data

```python
Nationality_percent = df_2.Nationality.value_counts(normalize = True)
ax = Nationality_percent.head(10).plot(kind = "bar", color = 'steelblue')
vals = ax.get_yticks()
ax.set_yticklabels(['{:,.2%}'.format(x) for x in vals])
print('\033[1m' + "Top 10 countries having the most number of players : " + '\033[0m')
plt.title("Percentages of Players corresponding to a Country")
plt.xlabel('Nationality')
plt.ylabel('Percentages')
plt.xticks(rotation = 45);
```

```python
df_2.groupby('Name')['Value_in_Euros_in_M'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', color = 'steelblue')
print('\033[1m' + "Top 10 Most valuable players : " + '\033[0m')
plt.title('Top 10 Valuable Players')
plt.xlabel('Players')
plt.ylabel('Value in Euros (Millions)')
plt.xticks(rotation = 45);
```

```python
top10_players_value = df_2.groupby('Club')['Value_in_Euros_in_M'].mean().sort_values(ascending = False).head(10)
top10_players_value.sort_values(ascending = True).plot(kind = 'barh', color = 'steelblue')
print('\033[1m' + "Top 10 Clubs with the most valuable players : " + '\033[0m')
plt.title('Top 10 Clubs with most Valuable Players')
plt.ylabel('Clubs')
plt.xlabel('Value in Euros (in Millions)')
plt.xticks(rotation = 0);
```

```python
df_2.groupby('Name')['Wage_in_Euros_in_M'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', color = 'steelblue')
print('\033[1m' + "Top 10 Highest Paid Players : " + '\033[0m')
plt.title('Top 10 Highest Paid Players')
plt.xlabel('Players')
plt.ylabel('Wages in Euros (Millions)')
plt.xticks(rotation = 45);
```
